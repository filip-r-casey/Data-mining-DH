```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
data = pd.read_csv('Final_MLB_Data.csv')
data.groupby('Name').Team.count().sort_values(ascending=False).head(10)
```

```python
data[data.Name == 'Yu Chang']
```

```python
data.sort_values('OFF', ascending=False).head(10)
```

```python
data.sort_values('DEF', ascending=False).head(10)
```

```python
player_off_hist = px.histogram(data, x='OFF', nbins=50, title='Offensive Score Distribution', height=700, width=1000)
player_off_hist.show()

player_def_hist = px.histogram(data, x='DEF', nbins=50, title='Defensive Score Distribution', height=700, width=1000)
player_def_hist.show()
```

```python
player = data[['Name','Team','OFF','DEF','Qualified DH']].copy()
player['Total Value'] = player.OFF + player.DEF
# player.sort_values('Total Value', ascending=False).head(10)
player_val_hist = px.histogram(player, x='Total Value', nbins=50, title='Total Value Distribution', color='Qualified DH', height=700, width=1000, barmode='overlay', opacity=0.7)
player_val_hist.show()
```

```python
player['is_full_DH'] = player['Qualified DH'].apply(lambda x: 1 if x == 'Yes' else 0)
team_df = player.groupby('Team').agg({'OFF':'sum', 'DEF':'sum', 'Total Value':'sum', 'is_full_DH':'max'}).reset_index()
team_df.sort_values('Total Value', ascending=False)
```

```python
## plot various distributions of offensive and defensive values

team_off_hist = px.histogram(team_df, x='OFF', nbins=50, title='Team Offensive Score Distribution', height=700, width=1000)
team_off_hist.show()

team_def_hist = px.histogram(team_df, x='DEF', nbins=50, title='Team Defensive Score Distribution', height=700, width=1000)
team_def_hist.show()

team_val_hist = px.histogram(team_df, x='Total Value', nbins=50, title='Team Total Value Distribution', height=700, width=1000)
team_val_hist.show()
```

```python
## box plot of team total value
team_val_box = px.box(team_df, x='Total Value', title='Team Total Value Box Plot', color='is_full_DH', height=700, width=1000, points='all')
team_val_box.show()
```
